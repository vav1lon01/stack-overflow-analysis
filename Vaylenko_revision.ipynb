```python
import pandas as pd
import numpy as np
```

Завдання 1. Підрахунок загальної кількості респондентів

Завдання 2. Аналіз повноти відповідей респондентів

```python
question = set(survey_results_schema['qname'])

columns = list(question.intersection(survey_results_public.columns))
dropna_columns = survey_results_public.dropna(subset = columns).shape[0]
print(f'The number of respondents that answered all questions is: {dropna_columns}')
```

Задание 4. Анализ удаленной работы

Задание 5. Определение популярности Python

```python
languages_columns = [col for col in survey_results_public.columns if 'Language' in col]
print(languages_columns)
```

```python
python_respondents = survey_results_public[
    survey_results_public['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)
]['ResponseId'].nunique()
total_respondents = survey_results_public['ResponseId'].nunique()
percent = python_respondents / total_respondents * 100
print(f"Percentage of respondents who want to work with Python: {percent:.2f}%")
```

Задание 6. Анализ путей обучения программированию

```python
languages_columns = [col for col in survey_results_public.columns if 'Learn' in col]
online_courses_count = survey_results_public[survey_results_public['LearnCode'].str.contains('Online Courses or Certification', na=False)].shape[0]
print(f'Online Courses or Certification : {online_courses_count}')

```

Задание 7. Географический анализ компенсации Python-разработчиков

```python
convertedCompYearly_columns = [col for col in survey_results_public.columns if 'ConvertedCompYearly' in col]
country_columns = [col for col in survey_results_public.columns if  'country' in col.lower()]

converted_dropna = survey_results_public.dropna(subset='ConvertedCompYearly')
avg_mean_country = converted_dropna[
    converted_dropna['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)
    ].groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).sort_values(by='median', ascending=False)
print(avg_mean_country)

```

Задание 8. Анализ образования наиболее оплачиваемых специалистов

```python
survey_results_public['EdLevel'] = survey_results_public['EdLevel'].fillna('Unknown')
sorted_converted = survey_results_public.sort_values(by='ConvertedCompYearly', ascending=False)
sorted_converted[['ResponseId', 'EdLevel', 'ConvertedCompYearly']].head(5)
```

Завдання 9. Аналіз популярності Python по віковим категоріям 



```python
age = set(survey_results_public['Age'])
print(age)
```

```python
puthon_respondet1 = survey_results_public[survey_results_public['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
age_python = puthon_respondet1.groupby('Age').size()
age_count_respondents = survey_results_public.groupby('Age').size()
percent_age_python = age_python / age_count_respondents * 100
print(percent_age_python)
```

Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників



```python
indastry_columns = [col for col in survey_results_public.columns if 'Industry' in col]
print(indastry_columns)
```

```python
survey_results_public['Industry'] = survey_results_public['Industry'].fillna('Unknown')
sort_remoute = survey_results_public[survey_results_public['RemoteWork'] == 'Remote']
quantile75_ind = sort_remoute[
    sort_remoute['ConvertedCompYearly'] >= sort_remoute['ConvertedCompYearly'].quantile(0.75)
].groupby('Industry')['ResponseId'].count().sort_values(ascending=False)

print(quantile75_ind)
```
